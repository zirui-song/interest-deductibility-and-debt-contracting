```python
import matplotlib.pyplot as plt
import numpy as np

# Firm labels
firms = ["Control Firm", "Threshold Firm", "Loss Firm"]

# Data (in millions)
ebitda = [60, 18, 8]
total_debt = [60, 60, 60]
interest_expense = [6, 5.4, 2.4]
excess_interest = [0, 0.6, 3.6]
excess_interest_pct = [0, 10, 60]  # percentages

# Slide 1: EBITDA vs. Total Debt (grouped bar chart)
x = np.arange(len(firms))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(8, 5))
rects1 = ax.bar(x - width/2, ebitda, width, label='EBITDA', color='skyblue')
rects2 = ax.bar(x + width/2, total_debt, width, label='Total Debt', color='salmon')

ax.set_ylabel('Amount ($ Millions)')
ax.set_title('EBITDA vs. Total Debt by Firm Type')
ax.set_xticks(x)
ax.set_xticklabels(firms)
ax.legend()

for rect in rects1 + rects2:
    height = rect.get_height()
    ax.annotate(f'{height}',
                xy=(rect.get_x() + rect.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')

plt.tight_layout()
plt.show()

# Slide 2: Interest Expense and Excess Interest (stacked bar chart)
fig, ax = plt.subplots(figsize=(8, 5))
p1 = ax.bar(firms, interest_expense, label='Interest Expense (Deductible)', color='lightgreen')
p2 = ax.bar(firms, excess_interest, bottom=interest_expense, label='Excess Interest', color='red')

ax.set_ylabel('Amount ($ Millions)')
ax.set_title('Interest Expense and Excess Interest')
ax.legend()

# Annotate each bar with values
for i, firm in enumerate(firms):
    ax.annotate(f'{interest_expense[i]}',
                xy=(i, interest_expense[i] / 2),
                ha='center', color='black')
    if excess_interest[i] > 0:
        ax.annotate(f'{excess_interest[i]}',
                    xy=(i, interest_expense[i] + excess_interest[i]/2),
                    ha='center', color='white')

plt.tight_layout()
plt.show()

# Slide 3: Excess Interest as a Percentage
fig, ax = plt.subplots(figsize=(8, 5))
bars = ax.bar(firms, excess_interest_pct, color=['blue', 'orange', 'red'])
ax.set_ylabel('Excess Interest (%)')
ax.set_title('Excess Interest Percentage by Firm Type')

# Annotate percentage on top of each bar
for bar in bars:
    height = bar.get_height()
    ax.annotate(f'{height}%',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # offset
                textcoords="offset points",
                ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
